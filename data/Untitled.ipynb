```python
import pandas as pd
import numpy as np
import math
```

```python
df1 = pd.read_csv("share-of-household-consumption-in-gdp-vs-gdp-per-capita.csv")
df2 = pd.read_csv("收入.csv")
df3 = pd.read_csv('gdp.csv')
df4 = pd.read_csv('unemployment_rate.csv')
```

```python
def data_search(country_code, year,df):
    year = str(year)
    matching_rows = df.loc[df["Country Code"] == country_code, year]
    
    if len(matching_rows) == 1:
        # 如果找到了一个匹配的值，返回这个值
        if math.isnan(matching_rows.item()):
            return None
        return int(matching_rows.item())
    else:
        # 如果没有找到，或者找到了多个匹配的值，返回None或其他指示值
        return None
```

```python
df1 = df1[['Entity', 'Year', 'consumption','GDP per capita']] #Share of household consumption in GDP
df2 = df2[['Entity', 'Year', 'GNI per capita']]
#df3 = df3[['Entity', 'Year', 'GDP']]
df4= df4[['Entity', 'Year', 'Unemployment']]
```

```python
final_common_df = df1\
    .merge(df2, on=['Entity', 'Year'], how='inner')\
    .merge(df4, on=['Entity', 'Year'], how='inner')
```

```python
#final_common_df['Life Expectancy'] = final_common_df.apply(lambda row: life_expectancy_search(row['Entity'], row['Year']), axis=1)

# 删除任何包含空值的行
final_common_df = final_common_df.dropna()

final_common_df
```

```python
final_common_df['Employment'] = 100 - final_common_df['Unemployment']
# 删除原来的 'Unemployment' 列
final_common_df.drop('Unemployment', axis=1, inplace=True)
```
